**OBSERVATION**



*   There are 50 rows and 6 coloumns in our dataset. Each rows contains some specifics informations of certain employee of the company.
*  The features in data sets are:


1.The id of the empolyee.

2.Their respective group  and age.


3.Their healthy eating and active lifestyle.

4.Their salary.







```python
plt.figure(figsize = (10,5))
sns.scatterplot(data = data , x = 'age' , y = 'salary' , color = 'blue' , edgecolor = 'linen' , alpha = 0.5)
plt.title("Scatterplot of Employee age v/s Employee Salary")
plt.xlabel("Employee age")
plt.ylabel("Employee Salary")
plt.show()
```

***OBSERVATION***



* We observe from our scatter plot that there is direct relation between age and salary of the employee.
* As age of employee increases salary of employee also increases so there are some outliers also present as well.



```python
plt.figure(figsize = (10,5))
sns.scatterplot(data = data , x ='healthy_eating' , y = 'active_lifestyle' , color = 'red' , edgecolor = 'linen' , alpha = 0.5)
plt.title("Scatterplot of Healty eating vs active lifestyle of employee")
plt.xlabel("Healthy eating")
plt.ylabel("Active lifestyle")
plt.show()
```

**OBSERVATION**


*We observe that the in genral there is relation between healthy eating and active lifestyle of the employee.

*As healthy eating increases active lifestyle increases.



```python
plt.figure(figsize = (10,5))
sns.scatterplot(data = data , x = 'healthy_eating' , y = 'salary' , color = 'k' , edgecolor = 'linen' , alpha = 0.5)
plt.title("Scatterplot of healthy eating vs salary")
plt.xlabel("Heatlthy eating")
plt.ylabel("Salary")
plt.show()
```

**OBSERVATION**



* From this scatter plot we observe that in genral there is relation between healthy eatinga and salary of the employee .

*As healthy eating increases salary also increases.



```python
plt.figure(figsize = (10,5))
sns.countplot(data = data , x = 'groups' , edgecolor = 'linen' , alpha = 0.7)
plt.title("Countplot of Groups of employee")
plt.xlabel("Groups")
plt.ylabel("Counts")
plt.show()
```

**OBSERVATION**

From this count plot we can observe that most employee belong to group A or group O with group A having maximum frequency.

```python
plt.figure(figsize = (10,5))
sns.histplot(data = data , x = 'salary' , color = 'red' , edgecolor = 'linen' , alpha = 0.7 , bins = 10)
plt.title("Histplot of salary of employee")
plt.xlabel("salary")
plt.ylabel("count")
plt.show()
```

**OBSERVATION**



*   From this Histplot we can observe that the salaries are not uniform which is not necessarily a discrepancy.

*   There are employees with salraies on either extriemities  of the histogram.



*   Based on the histogram , we can see that the count of the highest earing employee is very less.
